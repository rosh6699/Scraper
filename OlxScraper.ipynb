```python
import requests
from bs4 import BeautifulSoup
```

```python
r= requests.get("https://www.olx.in/items/q-Cars")
```

```python
c=r.content
```

```python
r
```

```python
soup=BeautifulSoup(c,'html.parser')
```

```python
print(soup.prettify())
```

```python
all= soup.find_all(attrs={"data-aut-id": "itemBox"})
```

```python
l=[]
for item in all:
    dict={}
    vehicles = item.find(attrs={"data-aut-id": "itemTitle"})
    dict["Vehicle"]= vehicles.text.split(',',1)[0]
    loc=item.find(attrs={"data-aut-id": "item-location"})
    dict["Location"]=loc.text.split(',',1)[0]
    pricekms = item.find(attrs={"data-aut-id": "itemPrice"})
    dict["Price"]= pricekms.text
    
    yp = item.find(attrs={"data-aut-id": "itemDetails"})
    try:
        dict["Kms"]= yp.text.split('-',1)[1]
        dict["Model"]=yp.text.split('-',1)[0]
        
    except:
        dict["Kms"]="-"
        dict["Model"]="-"
    
    #print(dict)
    l.append(dict)
```
